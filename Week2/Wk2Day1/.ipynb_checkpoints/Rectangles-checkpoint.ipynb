```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rectangle(x, left, right, height):
    w = 1000000000
    left_b = -left*w
    leftY = np.tanh(w*x + left_b)
    
    # ???
    right_b = -right*w
    rightY = np.tanh(w*x + right_b)
    
    return (leftY - rightY) * (height / 2)
```

```python
x = np.linspace(-5, 5, 1000)

w = 100
b = -100

y = rectangle(x, 2, 3, 5)
# Transition at -b/w

fig = plt.figure()
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 5000)
f = np.sin(x)

y = 0

num = 5000//40

for i in range(0, len(x) - num, num):
    height = (f[i] + f[i+num])/2
    
    y += rectangle(x, x[i], x[i+num], height)
    
fig = plt.figure()
plt.plot(x, f)
plt.plot(x, y)

plt.show()
```
